# Django简介和安装
使用Django这个web框架，搭建一个简单的量化监控平台。

pip install Django

django-admin --version

## 创建一个Django项目
django-admin startproject TradingMonitor

cd .\TradingMonitor\

python .\manage.py migrate    -表示创建或更新数据库模式。每当model源代码被改变后，如果我们要将其应用到数据库上，就要执行一次这个命令。

python .\manage.py createsuperuser -添加管理员账户。

python .\manage.py runserver -启动debugging模式
