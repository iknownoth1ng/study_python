## 1.1. <a id='toc1_1_'></a>[什么是单元测试](#toc0_)
单元测试是指一个自动化的测试：
- 用来验证一小段代码（单元）的正确性
- 可以快速执行
- 在独立的环境中执行

## 1.3. <a id='toc1_3_'></a>[编写第一个单元测试](#toc0_)
Python提供了unittest模块：
- 测试类继承unittest.TestCase
- 测试类的名字以Test开头
- 测试方法的名字以test_开头

```python
import unittest
from tests.basic.calculator import Calculator



class TestCalculator(unittest.TestCase):
    def test_add(self):
        # Setup
        cal = Calculator()
        expected_result = 10
        # Action
        actual_result = cal.add(2, 3, 5)

        # Assert
        self.assertEqual(actual_result, expected_result)


if __name__ == "__main__":
    # unittest.main()
    
    #* 如果在Jupyter下，请用以下方式运行单元测试
    unittest.main(argv=['first-arg-is-ignored'],exit=False)
```

## 1.4. <a id='toc1_4_'></a>[运行单元测试](#toc0_)
为了方便的运行测试
> pip install nose </br>
> pip install coverage </br>

运行一个测试文件：</br>
python -m unittest -v tests.withunittest.test_calculator

运行所有的测试文件：</br>
nosetests -v tests/withunittest/*

统计测试覆盖率： </br>
nosetests --with-coverage --cover-erase -v tests/withunittest/*
