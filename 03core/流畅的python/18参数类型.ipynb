# 特殊参数

## positional only parameters

```python
def timeV1(hour, minute, second):
    print(":".join([hour, minute, second]))


timeV1("9", "10", "7")
timeV1("9", minute="10", second="7")


def timeV2(hour, /, minute, second):
    print(":".join([hour, minute, second]))


timeV2("9", "19", "7")
timeV2("9", minute="10", second="7")
# timeV2(hour='9',minute='10',second='7') #! timeV2() got some positional-only arguments passed as keyword arguments: 'hour'
```

## keyword only parameters

```python
def timeV3(hour, *, minute, second):
    print(":".join([hour, minute, second]))


timeV3("9", minute="10", second="7")
timeV3("9", "19", "7")
```

```python
def timeV4(hour, /, minute, *, second):
    print(":".join([hour, minute, second]))


timeV4("9", minute="10", second="7")
timeV4("9", "10", second="7")
```
